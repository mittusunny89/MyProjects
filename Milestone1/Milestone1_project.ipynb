Tic Tac Toe game



```python

def display_board(my_list):
    print(' '+my_list[6]+' | '+my_list[7]+' | '+my_list[8]+' \n-----------\n '+my_list[3]+' | '+my_list[4]+' | '+my_list[5]+' \n-----------\n '+my_list[0]+' | '+my_list[1]+' | '+my_list[2]+' ')
    
def user_input(player):
    while True:
        pos = input(player + ', enter your position(1-9): ')
        if pos in tv and 1<= int(pos) <= 9:
            return pos
        elif pos not in tv and 1<= int(pos) <= 9:
            print('Sorry! position occupied. Please select a new one')
        elif int(pos)<1 or int(pos)>9:
            print('Sorry! Please select a valid position')
def change_value(pos,symbol):
    
        if pos == '1':
            tv[0] = symbol
        elif pos ==  '2':
            tv[1] = symbol
        elif pos == '3':
            tv[2] = symbol
        elif pos == '4':
            tv[3] = symbol            
        elif pos == '5':
            tv[4] = symbol
        elif pos == '6':
            tv[5] = symbol
        elif pos == '7':
            tv[6] = symbol
        elif pos == '8':
            tv[7] = symbol
        elif pos == '9':
            tv[8] = symbol
            
def win_check():
    if tv[0] == tv[1] == tv[2]:
        return True
    elif tv[0] == tv[3] == tv[6]:
        return True  
    elif tv[6] == tv[7] == tv[8]:
        return True 
    elif tv[8] == tv[5] == tv[2]:
        return True 
    elif tv[3] == tv[4] == tv[5]:
        return True 
    elif tv[1] == tv[4] == tv[7]:
        return True 
    elif tv[0] == tv[4] == tv[8]:
        return True 
    elif tv[2] == tv[4] == tv[6]:
        return True 
    
        
```

```python
tv= ['1','2','3','4','5','6','7','8','9']
game_on = True
isplayer1 = True
count = 0
while game_on:
    display_board(tv)
    if isplayer1 and count<9:
        current_pos = user_input('player1')
        change_value(current_pos,'X')
        if count>= 3 and win_check():
            print('Congrats!! player1 WON')
            game_on = False
        count+=1
    elif not isplayer1 and count<9:
        current_pos = user_input('player2')
        change_value(current_pos,'O')
        if count>= 3 and win_check():
             print('Congrats!! player2 WON')
             game_on = False
        count+=1

    else:
        print ("Game over!")
        break
    isplayer1 = not isplayer1

    
```
